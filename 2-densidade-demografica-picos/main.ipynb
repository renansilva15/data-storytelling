# 1. Definição

A densidade demográfica em Picos-PI desempenha um papel crucial no desenvolvimento urbano e socioeconômico. Compreender a distribuição dos domicílios e da população por setor censitário é essencial para o planejamento urbano e alocação eficiente de recursos públicos. A uniformidade ou disparidade nessa distribuição pode impactar acesso a serviços, infraestrutura e qualidade de vida dos cidadãos.

Hipótese 1: A distribuição dos domicílios por setor censitário em Picos é uniforme, refletindo uma ocupação equitativa do espaço urbano pela população.

Hipótese 2: A distribuição da população por setor censitário em Picos é heterogênea, com áreas de alta densidade demográfica concentradas em determinadas regiões da cidade

## Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Carrega os dados

```python
df_2010 = pd.read_csv(path_2010)
df_2010
```

```python
df_2022 = pd.read_csv(path_2022)
df_2022
```

# 3. Checagem

```python
df_2010.info()
```

```python
df_2022.info()
```

Como queremos verificar o número de domicílios por região precisamos remover os dados nulos da coluna **v0002** para os agregados de 2022 e para ter uma visão de como mudou de 2010 para cá podemos fazer isso também com a coluna **V001** dos agregados de 2010 como visto nos dicionários.

```python
df_2022 = df_2022.dropna(subset=['v0002'])
df_2010 = df_2010.dropna(subset=['V001'])
```

# 4. Análise

Primeiro para minimizar quaisquer erros de código setor(novos códigos, padrões, ...) filtramos ambos os dataframes por a cidade de Picos.

```python
df_2010_picos = df_2010[df_2010['Nome_do_municipio'] == 'PICOS'].reset_index()
df_2010_picos
```

```python
df_2022_picos = df_2022[df_2022['NM_MUN'] == 'Picos'].reset_index()
df_2022_picos
```

Agora vamos juntar as tabelas. Como podemos ver os códigos dos setores dos agregados de 2022 possuem esse "P" no final, imagino que seja de "Preliminar" mas não pesquisei a fundo. Iremos remover ele para juntarmos as duas tabelas apenas onde os códigos fore iguais.

```python
df_2022_picos['CD_SETOR'] = df_2022_picos['CD_SETOR'].str.replace(r'P$', '', regex=True)

df_2022_picos['CD_SETOR'].value_counts()
```

Agora mudamos o nome da coluna que desejamos juntar para corresponder a outra.

```python
df_2010_picos = df_2010_picos.rename(columns={'Cod_setor': 'CD_SETOR'})
```

Mudamos também o tipo.

```python
df_2022_picos['CD_SETOR'] = df_2022_picos['CD_SETOR'].astype('Int64')
```

Agora é só juntar

```python
df_picos_merged = pd.merge(df_2022_picos, df_2010_picos, on='CD_SETOR', how='inner')
```

```python
df_picos_merged.info()
```

Filtramos apenas os setores censitários da zona urbana (códigos 1, 2 e 3) pois os domicílios fora da área urbana foram agrupados em um único bairro.

```python
df_picos_urban = df_picos_merged[df_picos_merged['Situacao_setor'].isin([1, 2, 3])]
```

## Calcula o número de domicílios por bairro em 2010 e 2022

```python
households_by_urban_neighborhood2010 = df_picos_urban.groupby('Nome_do_bairro')['V001'].sum().sort_values(ascending=False)
households_by_urban_neighborhood2022 = df_picos_urban.groupby('Nome_do_bairro')['v0002'].sum().sort_values(ascending=False)
```

# 5. Visualização

## Mostra a distribuição da população por bairro

```python
plt.figure(figsize=(12, 8))
households_by_urban_neighborhood2010.plot(kind='bar', color='salmon')
plt.title('Distribuição de Domicílios por Bairro na Zona Urbana de Picos em 2010')
plt.xlabel('Bairro')
plt.ylabel('População')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
households_by_urban_neighborhood2022.plot(kind='bar', color='salmon')
plt.title('Distribuição de Domicílios por Bairro na Zona Urbana de Picos em 2022')
plt.xlabel('Bairro')
plt.ylabel('População')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A partir do trabalho é possível observar que a hipótese 2 está correta(a distribuição da população é heterôgenea). pode também se observar que proporção se manteve semelhante em um período de 12 anos com bairros como o Junco aumentando bastante em domicílios provavelmente por ser uma área perto da faculdade entre outras observações.
