# 1. Definição

A densidade demográfica em Picos-PI desempenha um papel crucial no desenvolvimento urbano e socioeconômico. Compreender a distribuição dos domicílios e da população por setor censitário é essencial para o planejamento urbano e alocação eficiente de recursos públicos. A uniformidade ou disparidade nessa distribuição pode impactar acesso a serviços, infraestrutura e qualidade de vida dos cidadãos.

Hipótese 1: A distribuição dos domicílios por setor censitário em Picos é uniforme, refletindo uma ocupação equitativa do espaço urbano pela população.

Hipótese 2: A distribuição da população por setor censitário em Picos é heterogênea, com áreas de alta densidade demográfica concentradas em determinadas regiões da cidade

```python
df = pd.read_csv(path, delimiter=";", encoding="latin1")
df
```

# 3. Checagem

```python
df.info()
```

```python
df_picos = df[df['Nome_do_municipio'] == 'PICOS'].reset_index()
df_picos
```

```python
df_picos.info()
```

Para respondermos as hipotéses antes levantadas iremos trabalhar com a coluna V002, segue abaixo o dicionário das colunas:

<img src="./assets/dicionario.png" alt="Dicionário das colunas">

## Remove os valores nulos de "V002"

```python
df_picos.dropna(subset="V002")
df_picos.info()
```

## Filtrar apenas os setores censitários da zona urbana (códigos 1, 2 e 3)

```python
df_picos_urban = df_picos[df_picos['Situacao_setor'].isin([1, 2, 3])]
```

## Calcula a população por bairro

```python
population_by_urban_neighborhood = df_picos_urban.groupby('Nome_do_bairro')['V002'].sum()
```

# 5. Visualização

## Mostra a distribuição da população por bairro

```python
plt.figure(figsize=(12, 8))
population_by_urban_neighborhood.plot(kind='bar', color='salmon')
plt.title('Distribuição da População por Bairro na Zona Urbana de Picos')
plt.xlabel('Bairro')
plt.ylabel('População')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para facilitar a leitura
plt.tight_layout()
plt.show()
```

A partir do trabalho é possível observar que a hipótese 2 está correta(a distribuição da população é heterôgenea). No trabalho foi utilizado apenasa zona urbana pois as demais estavam agrupadas em um único bairro genérico o que provocava um outlier. Deve ser levado em consideração que esses dados não contém todos os bairros pois datam de 2010 a 2013, outra coisa a ser levada em contada é que a população provavelmente mudou muito ao longo desse tempo. É provável que bairros mais próximos as universidades tenham aumentado em número de pessoas que o habitam.
