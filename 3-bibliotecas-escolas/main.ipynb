# 1. Definação

A importância das bibliotecas nas escola é inquestionável.
- Quantos porcento das escolas do Brasil possuem bibliotecas?
- Quais regiões possuem menos bibliotecas?
- Quais estados possuem menos bibliotecas?
- Qual a porcentagem de bibliotecas por o número de escolas de cada estado?

# 2. Obtenção

Para esta análise foram utilizados os dados do Censo Escolar de 2023 que podem ser baixados [aqui](https://download.inep.gov.br/dados_abertos/microdados_censo_escolar_2023.zip).

Depois de baixar os dados extrai o arquivo "microdados_ed_basica_2023.csv", crie uma pasta chamada datasets no diretório deste notebook e coloque o arquivo dentro.

```python
path = "./datasets/microdados_ed_basica_2023.csv"
```

# 3. Checagem

## Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(path, delimiter=';', encoding='latin1')
df
```

```python
df.info()
```

```python
df.describe()
```

# 4. Análise

Obtém o total de escolas:

```python
total_schools = len(df)
total_schools
```

Obtém as escolas que possuem livraria(s):

```python
schools_with_libraries = df[df['IN_BIBLIOTECA'] == 1]
schools_with_libraries
```

Calcula a porcentagem de escolas que possuem livrarias:

```python
percentage_schools_with_libraries = (len(schools_with_libraries) / total_schools) * 100
percentage_schools_with_libraries
```

Calcula o número de bibliotecas por escola por região:

```python
label_map = {0: 'WITHOUT_LIBRARY', 1: 'WITH_LIBRARY'}
df['IN_BIBLIOTECA_LABEL'] = df['IN_BIBLIOTECA'].map(label_map)

df_libraries_per_region = df.groupby('NO_REGIAO')['IN_BIBLIOTECA_LABEL'].value_counts().unstack().fillna(0)

df_libraries_per_region['TOTAL_SCHOOLS'] = df_libraries_per_region.sum(axis=1)

df_libraries_per_region_sorted = df_libraries_per_region.sort_values(by='WITH_LIBRARY', ascending=False)
df_libraries_per_region_sorted.drop(columns=['TOTAL_SCHOOLS', 'WITHOUT_LIBRARY'], inplace=True)

df_libraries_per_region_sorted
```

Obtém as livrarias agrupadas por estado:

```python
libraries_by_state = df.groupby('SG_UF')['IN_BIBLIOTECA'].sum().sort_values(ascending=False)
libraries_by_state
```

Obtém a porcentagem de bibliotecas por o número de escolas de cada estado:

```python
schools_libraries_by_state = df.groupby('SG_UF').agg({'IN_BIBLIOTECA': 'sum', 'CO_ENTIDADE': 'count'})
schools_libraries_by_state.columns = ['NUM_LIBRARIES', 'NUM_SCHOOLS']
  
schools_libraries_by_state['PERCENTAGE_LIBRARIES'] = (schools_libraries_by_state['NUM_LIBRARIES'] / schools_libraries_by_state['NUM_SCHOOLS']) * 100

schools_libraries_by_state_sorted = schools_libraries_by_state.sort_values(by='NUM_LIBRARIES', ascending=False)
schools_libraries_by_state_sorted
```

```python
libraries_per_school_percentage = (df.groupby('SG_UF')['IN_BIBLIOTECA'].sum() / df.groupby('SG_UF')['CO_ENTIDADE'].count()) * 100
libraries_per_school_percentage_sorted = libraries_per_school_percentage.sort_values(ascending=False)
libraries_per_school_percentage_sorted
```

# 5. Visualização

## Quantos porcento das escolas do Brasil possuem bibliotecas?

```python
plt.figure(figsize=(8, 6))
labels = ['Com livrarias', 'Sem livrarias']
sizes = [percentage_schools_with_libraries, 100 - percentage_schools_with_libraries]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem de escolas com livrarias')
plt.show()
```

Aqui podemos ver que apenas pouco mais de um terço das escolas do Brasil possuem bibliotecas.

## Quais regiões possuem menos bibliotecas?

```python
plt.figure(figsize=(10, 6))
df_libraries_per_region_sorted.plot(kind='bar')
plt.title('Número de Bibliotecas por Região')
plt.xlabel('Região')
plt.ylabel('Número de Bibliotecas')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Aqui podemos ver o número de de bibliotecas por região mas não levamos em conta quantas escolas são por cada região.

## Quais estados possuem menos bibliotecas?

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=libraries_by_state.index, y=libraries_by_state.values)
plt.title('Número de Bibliotecas por estado')
plt.xlabel('Estado')
plt.ylabel('Número de bibliotecas')
plt.xticks(rotation=45)
plt.show()
```

Aqui podemos ver que apesar de no gráfico anterior mostrar que o sul tem o maior número de bibliotecas, Minas Gerais é o estado que detem o maior número de bibliotecas. Porém não verificamos por número de escolas por estado.

## Qual a porcentagem de bibliotecas por o número de escolas de cada estado?

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=schools_libraries_by_state_sorted.index, y='NUM_SCHOOLS', data=schools_libraries_by_state_sorted, color='skyblue', label='Escolas')
sns.barplot(x=schools_libraries_by_state_sorted.index, y='NUM_LIBRARIES', data=schools_libraries_by_state_sorted, color='salmon', label='Bibliotecas')

plt.title('Número de bibliotecas por escola em cada Estado')
plt.xlabel('Estado')
plt.ylabel('Escolas e Bibliotecas')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Aqui podemos ver o número de bibliotecas por escola em cada estado complementando a pergunta anterior.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=libraries_per_school_percentage_sorted.index, y=libraries_per_school_percentage_sorted.values, color='salmon')
plt.title('Porcentagem de bibliotecas por número de escolas por estado')
plt.xlabel('State')
plt.ylabel('Percentage of Libraries')
plt.xticks(rotation=45)
plt.show()
```

Aqui podemos ver a porcentagem de bibliotecas por escola em cada estado. Observamos que no estado de Goiás mais de 50% das suas escolas possuem bibliotecas enquanto o estado do Acre que é o que tem menos escolas também é o que tem a menor porcentagem.

Com essa análise é possível fazer várias observações, sendo algumas delas:
- O estado de Minas Gerais detem a grande maioria das biblotecas da sua região.
- O Brasil possúi apenas 3 estados com 50%+ de bibliotecas para suas escolas o que já não é um número bom.
- O estado do Acre apresenta um número bem menor do que os demais estados o que pode levar a problemas que pode ser observados em outros estudos.
